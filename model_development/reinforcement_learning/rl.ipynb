#### 1. Import Libraries

```python
import warnings
warnings.filterwarnings("ignore")

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

#### 2. Define Environment

```python
ENVIRONMENT = "Loan Approval System"

print(f"Environment : {ENVIRONMENT}")
```

#### 3. Define States

```python
states = {
    0: "Low Risk",
    1: "Medium Risk",
    2: "High Risk"
}

n_states = len(states)
print(states)
```

#### 4. Define Actions

```python
actions = {
    0: "Approve",
    1: "Review",
    2: "Reject"
}

n_actions = len(actions)
print(actions)
```

#### 5. Define Reward Matrix

```python
reward_matrix = np.array([
    # Approve, Review, Reject
    [100, 40, -20],    # Low Risk
    [20, 60, 10],    # Medium Risk
    [-300, 20, 100]     # High Risk
])

reward_df = pd.DataFrame(
    reward_matrix,
    index=states.values(),
    columns=actions.values()
)

display(reward_df)
```

#### 6. Initialize Q Table

```python
q_table = np.zeros(
    (n_states, n_actions)
)

q_df = pd.DataFrame(
    q_table,
    index=states.values(),
    columns=actions.values()
)

display(q_df)
```

#### 7. Hyperparameters

```python
ALPHA = 0.10 # Learning Rate
GAMMA = 0.90 # Discount Factor
EPSILON = 1.00 # Exploration Rate
EPSILON_DECAY = 0.995 
MIN_EPSILON = 0.01 
EPISODES = 5000 
```

#### 8. Environment Function

```python
def step(state, action):

    # Get reward
    reward = reward_matrix[state, action]

    # Simulate next customer (temporary)
    next_state = np.random.randint(0, n_states)

    return next_state, reward
```

#### 9. Epsilon Greedy Policy

```python
def choose_action(state, epsilon):

    # Exploration
    if np.random.rand() < epsilon:

        return np.random.randint(n_actions)

    # Exploitation
    return np.argmax(q_table[state])
```

#### 10. Update Q Table

```python
def update_q_table(
    state,
    action,
    reward,
    next_state
):

    old_value = q_table[state, action]

    next_max = np.max(
        q_table[next_state]
    )

    new_value = old_value + ALPHA * (
        reward +
        GAMMA * next_max -
        old_value
    )

    q_table[state, action] = new_value
```

#### 11. Training Function

```python
def train():

    epsilon = EPSILON
    reward_history = []

    for episode in range(EPISODES):

        # Random starting customer
        state = np.random.randint(n_states)

        total_reward = 0

        # One decision per episode
        action = choose_action(
            state,
            epsilon
        )

        next_state, reward = step(
            state,
            action
        )

        update_q_table(
            state,
            action,
            reward,
            next_state
        )

        total_reward += reward

        reward_history.append(
            total_reward
        )

        # Reduce exploration gradually
        epsilon = max(
            MIN_EPSILON,
            epsilon * EPSILON_DECAY
        )

    return reward_history
```

```python
# 12. Final Q Table
print("\nFinal Q Table")
q_df = pd.DataFrame(
    np.round(q_table, 2),
    index=states.values(),
    columns=actions.values()
)

print(q_df)
```

```python
# 13. Learned Policy
print("\nLearned Policy")

for state in range(n_states):

    best_action = np.argmax(q_table[state])

    print(
        f"{states[state]} "
        f"-> "
        f"{actions[best_action]}"
    )
```

```python
def evaluate_agent():

    print("AGENT DECISIONS")

    for state in range(n_states):

        action = np.argmax(q_table[state])

        reward = reward_matrix[state, action]

        print(
            f"{states[state]:15}"
            f" -> "
            f"{actions[action]:10}"
            f" Reward = {reward}"
        )

evaluate_agent()
```
